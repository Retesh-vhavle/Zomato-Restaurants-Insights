# replacing new ,/5 and - with nan

```python
def handlerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate']= df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.info()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace = True)
```

```python
def handlecost(value):
    value = str(value)
    if ',' in value :
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['approx_cost(for two people)']= df['approx_cost(for two people)'].apply(handlecost)
df['approx_cost(for two people)'].unique()
```
